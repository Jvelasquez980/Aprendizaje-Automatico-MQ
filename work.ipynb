```python
from pathlib import Path

import cv2 as cv



# Carpeta de entrada con las imágenes originales

data_root = Path("data") / "Blood cell Cancer [ALL]"



# Carpeta de salida donde se guardarán las imágenes en escala de grises

output_root = Path("data_grayscale") / "Blood cell Cancer [ALL]"

output_root.mkdir(parents=True, exist_ok=True)



image_exts = {".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff"}



num_convertidas = 0

num_fallidas = 0



for img_path in data_root.rglob("*"):

    if img_path.suffix.lower() not in image_exts:

        continue



    # Ruta relativa respecto a la carpeta de datos original

    rel_path = img_path.relative_to(data_root)

    out_path = output_root / rel_path

    out_path.parent.mkdir(parents=True, exist_ok=True)



    img = cv.imread(str(img_path))

    if img is None:

        print(f"No se pudo leer: {img_path}")

        num_fallidas += 1

        continue



    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)



    if cv.imwrite(str(out_path), gray):

        num_convertidas += 1

    else:

        print(f"No se pudo guardar: {out_path}")

        num_fallidas += 1



print(f"Conversión completada. Imágenes convertidas: {num_convertidas}, fallidas: {num_fallidas}.")

print(f"Las imágenes en gris están en: {output_root}")
```

```python
from pathlib import Path

import cv2 as cv



# Carpeta de entrada con las imágenes en escala de grises

gray_root = Path("data_grayscale") / "Blood cell Cancer [ALL]"



# Carpeta de salida para las imágenes con histograma ecualizado

eq_root = Path("data_equalized") / "Blood cell Cancer [ALL]"

eq_root.mkdir(parents=True, exist_ok=True)



image_exts = {".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff"}



num_ecualizadas = 0

num_fallidas = 0



for img_path in gray_root.rglob("*"):

    if img_path.suffix.lower() not in image_exts:

        continue



    # Ruta relativa respecto a la carpeta de entrada

    rel_path = img_path.relative_to(gray_root)

    out_path = eq_root / rel_path

    out_path.parent.mkdir(parents=True, exist_ok=True)



    # Leemos en modo escala de grises por seguridad

    img_gray = cv.imread(str(img_path), cv.IMREAD_GRAYSCALE)

    if img_gray is None:

        print(f"No se pudo leer: {img_path}")

        num_fallidas += 1

        continue



    # Ecualización de histograma

    img_eq = cv.equalizeHist(img_gray)



    if cv.imwrite(str(out_path), img_eq):

        num_ecualizadas += 1

    else:

        print(f"No se pudo guardar: {out_path}")

        num_fallidas += 1



print(f"Ecualización completada. Imágenes ecualizadas: {num_ecualizadas}, fallidas: {num_fallidas}.")

print(f"Las imágenes ecualizadas están en: {eq_root}")
```
